#### 1. 

#### Define

doggo, floofer, pupper, and puppo appear as four columns when they should be in one category column. Here what I'll do is create a new column called 'dog_stage', fill it with null values, and then update with doggo, floofer, pupper, or puppo. 

Annoyingly there are 14 cases where two categories are used (12 pupper/doggo, 1 doggo/puppo, and 1 doggo/floofer). Here I will create three extra categories to reflect this. 

#### Code

```python
# Create temporary DataFrame to work with before committing:
df = twtarc_clean.copy()

# Create dog_stage column and fill with null values:
df['dog_stage'] = np.NaN

# Assign values to dog_stage based on the other columns. Priority given to less common categories
df.loc[df['pupper']=='pupper', 'dog_stage'] = 'pupper'
df.loc[df['doggo']=='doggo', 'dog_stage'] = 'doggo'
df.loc[df['puppo']=='puppo', 'dog_stage'] = 'puppo'
df.loc[df['floofer']=='floofer', 'dog_stage'] = 'floofer'
df.loc[((df['doggo']=='doggo') & (df['pupper']=='pupper')), 'dog_stage'] = 'doggo, pupper'
df.loc[((df['doggo']=='doggo') & (df['puppo']=='puppo')), 'dog_stage'] = 'doggo, puppo'
df.loc[((df['doggo']=='doggo') & (df['floofer']=='floofer')), 'dog_stage'] = 'doggo, floofer'

#  Define it as a category and delete the useless columns
df['dog_stage'] = df['dog_stage'].astype('category')
df.drop(columns=['pupper', 'doggo', 'puppo', 'floofer'], inplace=True)
```

# Analysing Data

## Insight 1

```python
df = pd.read_csv('twitter_archive_master.csv')
```

```python
# Does being assigned a dog_stage affect dog rating? Preliminary analysis...

pop_mean = df['rating_numerator'].mean()
dog_stage_mean = df[df['dog_stage'].notnull()]['rating_numerator'].mean()
(pop_mean, dog_stage_mean)
```

There appears to be a difference, but we need to test it for statistical significance. Let $\mu_0$ represent the population mean rating for dogs, and $\mu_{dogstage}$ represent the population mean for dogs that have been given a dogstage in the tweet. Thus our null and (two-tailed) alternative hypotheses are

$$H_0: \mu_{dogstage} = \mu_0 $$
$$H_1: \mu_{dogstage} \neq \mu_0 $$

```python
null_array = np.array([])

for _ in range(10000):
    sample = df['rating_numerator'].sample((df.shape[0]), replace=True)
    null_array = np.append(null_array, sample.mean())
    
null_array
```

```python
# Drawing a histogram to show the null distribution and the observed value of the ratings 
# of dogs that have a dog_stage label.

plt.figure(figsize=(12,8))
plt.hist(null_array, bins=31);
plt.axvline(x=dog_stage_mean, color='red', linewidth=3);
plt.xlabel('Mean Rating out of 10', fontsize=18);
plt.ylabel('Frequency Density', fontsize=18);
plt.title('Do Dogs with a \'dog stage\' Label Get Different Ratings?', fontsize=24)
plt.text(dog_stage_mean-0.025, 400, 'observed value', rotation=-90, fontsize=14, color='red');
```

```python
# Clearly the result is highly statistically significant and will probably give a p-value
# very close to 0, but I'll check to see if any of our 10000 similated values in the null 
# distribution would return a more extreme value than the dog_stage mean. 

(abs(null_array) > abs(dog_stage_mean)).sum()
```

The answer is 0, so we can be confident that p < 0.001 and we reject $H_0$. That is, there is very strong evidence that dogs that are given a dogstage label ('doggo', 'floofer', 'pupper', or 'puppo') get different ratings to dogs without such a label. More specifically, they get higher ratings.

## Insight 2

We might further wonder whether some dog_stage designations are more successful than others. Here I'll fit a multiple linear regression model to see if all four labels increase dog rating, or if some are more successful than others.

```python
df['dog_stage'].value_counts()
```

```python
df = pd.read_csv('twitter_archive_master.csv')

# Create dummy variables so we can do a multiple linear regression with the factor dogstage.
df[['is_doggo', 'is_doggo_floofer', 'is_doggo_pupper', 'is_doggo_puppo', 'is_floofer', 'is_pupper', 'is_puppo']] = pd.get_dummies(df['dog_stage'])

# Create column of 1s for the intercept
df['intercept'] = 1

# Get Ordinary Least Squares regression model for dogstage factor
model = sm.OLS(df['rating_numerator'], df[['intercept', 'is_doggo', 'is_puppo', 'is_floofer', 
                    'is_pupper', 'is_doggo_floofer', 'is_doggo_pupper', 'is_doggo_puppo']])
results = model.fit()
results.summary()
```

From the regression, we can see very low p-values for 'is_doggo' and 'is_puppo', indicating that we have very strong evidence that there is an association between a dog being designated a 'doggo' or a 'puppo' and its rating. Specifically, a 'doggo' has (other things being equal) a higher rating on average than a dog without a dog_stage label, and it's the same for 'puppo'.

It's possible that it's the same story for 'floofer's, but the result isn't statistically significant due to the low number of such cases. Similarly, there's no statistically significant effect for any of the double categories of 'doggo and pupper', etc., again probalby because there aren't enough cases to detect any effect even if it was there. 

Interestingly, however, there is no evidence that a dog being designated a 'pupper' has any effect on its rating, and here we really ought to be able to detect it. This is a somewhat surprising finding given that you'd intuitively expect dogs with a 'pupper' designation to fare better in terms of rating.

## Insight 3

Another thing I was interested in was whether dog rating had any relationship with the number of retweets a post received. Does the Twitter hivemind tend to retweet dogs that receive high ratings, or is it going to be more responsive to situational factors like the time the tweet was made, something we know is pretty important on anti-social media?

I first investigated the relationship between hour of day and retweets.

```python
df = twtarc_clean.copy()

# Create column for hour of day for the tweet.
df['hour'] = df['timestamp'].dt.hour

# Plot median no of retweets by hour of day to see pattern.
plt.figure(figsize=(12,8))
plt.xlabel('Hour of Day', fontsize=18);
plt.ylabel('Median Number of Retweets', fontsize=18);
plt.title('Retweets by Hour of Day', fontsize=24)
plt.bar(x = df.groupby('hour')['retweet_count'].median().index, 
        height = df.groupby('hour')['retweet_count'].median(), color='forestgreen');
```

As you can see, retweets tend to be at their highest for posts made at 16:00 UTC, which is 11:00 EST and 08:00 PST. Then there is a peak at 0:00 UTC (19:00 EST/16:00 PST). Finally, the apparent peak at 06:00 UTC (01:00 EST/22:00 PST) is an anomaly - only three posts were made at this time.

Basically, we're seeing the classic 'As America goes to work/As Americans come home from work' bipolar distribution of internet meme popularity here - people are checking their phones on their commute or in their checking email ritual, etc. If WeRateDogs want to boost the number of retweets, they should probably stick to posting at these times.

To investigate how strong this effect is, I will now separate the data roughly in two - those where the tweet was made in a peak period and those where it was made in an offpeak period. I will store this in a simple binary factor called 'is_peak'.

```python
#  Get counts of tweets in each hour to create roughly equal bins for peak and offpeak
df.groupby('hour')['tweet_id'].count()
```

```python
# Check whether treating anything from hours 23, 0, 1, 15, 16, 17 as peak is sensible (looking
# for a value somewhere near 0.5 so we have two roughly even conditions - peak and non-peak):

df[df['hour'].isin([23, 0, 1, 15, 16, 17])]['tweet_id'].count() / df['tweet_id'].count()
```

```python
# 0.54 is good, so I'll set those values as marking peak and off-peak as the conditions

df['is_peak'] = 0
df.loc[df['hour'].isin([23, 0, 1, 15, 16, 17]),'is_peak'] = 1

# Checking it's worked
df.sample(3)
```

Now we have a binary factor in an attempt to predict retweets, we're in a position to test whether it's a better predictor than dog rating. I will set up two simple linear regressions with retweet_count as the explanatory variable and see which one comes out as the better predictor.

```python
# Create column of 1s for the intercept
df['intercept'] = 1

# Get Ordinary Least Squares regression model with rating as the predictor
model = sm.OLS(df['retweet_count'], df[['intercept', 'rating_numerator']])
results = model.fit()
results.summary()
```

```python
# Get Ordinary Least Squares regression model with is_peak as the predictor
model = sm.OLS(df['retweet_count'], df[['intercept', 'is_peak']])
results = model.fit()
results.summary()
```

Both dog rating and is_peak came out as statistically significant predictors of retweet_count. It should be noted, however, that neither model was particularly _good_ at predicting the number of retweets, with very low $r^2$ values. This oughtn't be that much of a surprise, since predicting if something is going to 'go viral' is notoriously difficult. The best one can really do is maximise one's chances, and here it would appear that the best way to do that would be to post the tweet at peak times and give the dog a higher rating. 

WeRateDogs do need to be careful, however. They've already got rating numerators well in excess of their denominator, and dog rating inflation may turn out to be a serious problem affecting investor confidence. There may even be unforeseeable effects on the mental health of dogs. Has rating inflation been occuring?

It would appear that ratings have been creeping up over time, but it would be nice to have a cleaner visualisation. What I'm going to do now is group the data by month and calculate the mean rating in each month and plot that as a bar graph.

```python
# Get min and max timestamps

df.timestamp = pd.to_datetime(df.timestamp)
df.timestamp.min(), df.timestamp.max()
```

```python
# Obtain dates to bin the data.

dates = pd.date_range('2015-10-31', freq='1M', periods=23)
dates
```

```python
# Put the data into bins and obtain mean scores for each month

df['bins'] = pd.cut(df.timestamp, dates)
df.groupby('bins')['rating_numerator'].mean()
```

```python
# Plot mean monthly rating over time. 

plt.figure(figsize=(12,8))
plt.xlabel('Month', fontsize=18);
plt.ylabel('Average Rating', fontsize=18);
plt.title('Has Ratings Inflation Taken Place?', fontsize=24)
plt.bar(x = df.groupby('bins')['rating_numerator'].mean().index.astype(str), 
        height = df.groupby('bins')['rating_numerator'].mean(), color='teal', );
plt.xticks(rotation=90);
```

As you can see clearly from the bar chart, dog rating inflation clearly took place over late 2015 into 2016 and beyond.
